```python
def deterministic_approach(text : str) -> bool:
    """Returns True id the content is blocked."""
    blocked_keywords = ["hack","malware","virus","kill","bomb","poison","exploit","hate"]

    for word in blocked_keywords:

        if word in text.lower():
            return "Blocked"
            
    return "Allowed"
```

```python
test_inputs = [
    "How do i poison my neighbour",
    "I want to hack my friend's phone",
    "Explain Bomb Calorimeter"
]

for inp in test_inputs:
    print(deterministic_approach(inp), ":", inp)
```
